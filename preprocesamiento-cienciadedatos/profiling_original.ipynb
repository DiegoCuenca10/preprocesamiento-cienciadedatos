```python
import cProfile      # Se importa cProfile 

def es_primo(n):
    if n < 2:           # Menor que 2 no es primo 
        return False
    for i in range(2, int(n**0.5)+1):   # Verifica los divisores hasta la raiz cuadrada 
        if n % i == 0:                  # Si es divisibles entonces no es primo 
            return False
    return True

def obtener_primos():
    return [n for n in range(1, 100000) if es_primo(n)]  

# Guardar profiling en archivo
cProfile.run("obtener_primos()", filename="profiling_original.txt") # Ejecuta el profiling 

print("profiling_original.txt generado correctamente.")
```
